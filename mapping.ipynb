```python
import pandas as pd
import os
```

```python
zibbase = 'zibs2020'
zibspath = os.path.join(zibbase, 'mappable')
zibs = os.listdir(zibspath)
zibs
```

```python
zib = 'Nl.zorg.Problem-v4.4(2020EN).xlsx'
print(zib)
zibpath = os.path.join(zibspath, zib)
zibdf = pd.read_excel(zibpath)
zibdf
```

```python
xlmbase = 'xtehr'
xlmspath = os.path.join(xlmbase, 'mappable')
xlms = os.listdir(xlmspath)
xlms
```

```python
xlm = 'Condition.xlsx'
print( xlm)
xlmpath = os.path.join(xlmspath, xlm)
xlmdf = pd.read_excel(xlmpath)
xlmdf
```

```python
mapbase = 'maps'
mapspath = os.path.join(mapbase, 'xtehr2zibs')
maps = os.listdir(mapspath)
maps
```

```python
file = 'Condition.xlsx'
print(file)
mappath = os.path.join(mapspath, file)
mapdf = pd.read_excel(mappath)
mapdf
```

```python
mapping = mapdf[['xtehr', 'zib']].merge(zibdf.add_suffix('_zib'), left_on='zib', right_on='path_zib', how='outer').merge(xlmdf.add_suffix('_xtehr'), left_on='xtehr', right_on='path_xtehr', how='outer')
mapping
```

```python
shortmap = mapping[['zib', 'xtehr', 'type_zib','type_xtehr', 'card._zib', 'card._xtehr']]
shortmap
```

```python
filepath = file.replace('.xlsx', '-full.xlsx')
mappath = os.path.join(mapspath, filepath)
mapping.to_excel(mappath, index=False)
```

```python
filepath = file.replace('.xlsx', '.csv')
mappath = os.path.join(mapspath, filepath)
mapping.to_csv(mappath, index=False)
```

```python
from IPython.display import Markdown
import numpy as np

content = "# " + mapping.iat[0, 0][4:] + "\n\n"
content += shortmap.fillna('').to_markdown(index=False) + "\n\n"
for i, row in mapping.iterrows():
    # print(i, row)
    content += "\n\n"
    if row.iloc[0] is not np.nan:
        content += "## " + str(row.iloc[0]) + "\n\n"
    else:
        content += "## zib: " + str(row.iloc[1]) + "\n\n"
    content += "### Table" + "\n\n"
    content += "| attribute | value |" + "\n" + "|---|---|" + "\n"
    for j, att in row.fillna('')[0:2].items():
        content += "| " + str(j) + " | " + str(att) + " |" + "\n"
    for j, att in row.fillna('')[2:].sort_index().items():
        content += "| " + str(j) + " | " + str(att) + " |" + "\n"    
    content += "\n### Comments" + "\n\n"
with open(os.path.join(mapspath, file.replace('.xlsx', '.md')), 'w') as f:
    f.write(content)
Markdown(content)
```
