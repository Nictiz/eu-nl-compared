```python
import pandas as pd
import os
import unicodedata as ud
```

```python
# zibbase = 'zibs2020'
zibbase = 'zibs2024'
zibspath = os.path.join(zibbase, 'mappable')
zibs = os.listdir(zibspath)
[(i, zib) for i, zib in enumerate(zibs)]
```

```python
zib = zibs[4]
print(zib)
zibpath = os.path.join(zibspath, zib)
zibdf = pd.read_excel(zibpath)
zibdf['path'] = zibdf['path'].apply(
    lambda x: ud.normalize("NFKC", str(x)).replace("\xa0", " ").strip() if pd.notna(x) else x
)
zibdf
# only for zib2024 condition and diagnosis
# zib = zibs[1]
# print( zib)
# zibpath = os.path.join(zibspath, zib)
# zibdf2 = pd.read_excel(zibpath)
# zibdf = pd.concat([zibdf, zibdf2], ignore_index=True)
# zibdf
```

```python
xlmbase = 'xtehr'
xlmspath = os.path.join(xlmbase, 'mappable')
xlms = os.listdir(xlmspath)
[(i, xlm) for i, xlm in enumerate(xlms)]
```

```python
xlm = xlms[9]
print( xlm)
xlmpath = os.path.join(xlmspath, xlm)
xlmdf = pd.read_excel(xlmpath)
# Normalize + strip (for non-breaking spaces, Unicode issues)
xlmdf['path'] = xlmdf['path'].apply(
    lambda x: ud.normalize("NFKC", str(x)).replace("\xa0", " ").strip() if pd.notna(x) else x
)
xlmdf
# only for device use and organization
# xlm = xlms[7]
# print( xlm)
# xlmpath = os.path.join(xlmspath, xlm)
# xlmdf2 = pd.read_excel(xlmpath)
# xlmdf = pd.concat([xlmdf, xlmdf2], ignore_index=True)
# xlmdf
```

```python
mapbase = 'maps'
maps = os.listdir(mapbase)
[(i, map) for i, map in enumerate(maps)]
```

```python
file = maps[8]
print(file)
mappath = os.path.join(mapbase, file)
print(mappath)
mapdf = pd.read_excel(mappath)
# Normalize + strip (for non-breaking spaces, Unicode issues)
mapdf[['zib','xtehr']] = mapdf[['zib','xtehr']].map(
    lambda x: ud.normalize("NFKC", str(x)).replace("\xa0", " ").strip() if pd.notna(x) else x
)
mapdf
```

```python
mapping = mapdf[['xtehr', 'zib']].merge(zibdf.add_suffix('_zib'), left_on='zib', right_on='path_zib', how='left').merge(xlmdf.add_suffix('_xtehr'), left_on='xtehr', right_on='path_xtehr', how='left')
mapping
```

```python
shortmap = mapping[['zib', 'xtehr', 'type_zib','type_xtehr', 'card._zib', 'card._xtehr']]
shortmap
```

```python
mapsoutpath = os.path.join(mapbase, 'xtehr2zibs2024')
filepath = file.replace('.xlsx', '-full.xlsx')
mappath = os.path.join(mapsoutpath, filepath)
mapping.to_excel(mappath, index=False)
```

```python
filepath = file.replace('.xlsx', '.csv')
mappath = os.path.join(mapsoutpath, filepath)
mapping.to_csv(mappath, index=False)
```

```python
from IPython.display import Markdown
import numpy as np

content = "# " + mapping.iat[0, 0][4:] + "\n\n"

# shortmap: linefeeds naar <br> vlak vr to_markdown
shortmap_md = shortmap.fillna('').map(
    lambda x: x.replace("\r\n", "<br>").replace("\n", "<br>") if isinstance(x, str) else x
)
content += shortmap_md.to_markdown(index=False) + "\n\n"

for i, row in mapping.iterrows():
    content += "\n\n"
    if row.iloc[0] is not np.nan:  # desgewenst: gebruik pd.notna(row.iloc[0])
        content += "## " + str(row.iloc[0]) + "\n\n"
    else:
        content += "## zib: " + str(row.iloc[1]) + "\n\n"

    content += "### Table" + "\n\n"
    content += "| attribute | value |\n|---|---|\n"

    # Eerste 2 kolommen
    for j, att in row.fillna('')[0:2].items():
        content += "| " + str(j) + " | " + (att.replace("\r\n","<br>").replace("\n","<br>") if isinstance(att, str) else str(att)) + " |\n" 

    # Rest, gesorteerd
    for j, att in row.fillna('')[2:].sort_index().items():
        if len(str(att)) > 0:
            content += "| " + str(j) + " | " + (att.replace("\r\n","<br>").replace("\n","<br>") if isinstance(att, str) else str(att)) + " |\n"

    content += "\n### Comments\n\n"

with open(os.path.join(mapsoutpath, file.replace('.xlsx', '.md')), 'w') as f:
    f.write(content)

# PAS OP: deze overschrijft de evt. handmatige comments; alleen bij eerste keer mapping
# with open(os.path.join(mapsoutpath, file.replace('.xlsx', '-comments.md')), 'w') as f:
#     f.write(content)

Markdown(content)
```
